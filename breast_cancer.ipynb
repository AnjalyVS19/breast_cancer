```python
x=df.drop([6,10],axis=1)
x
```

```python
y=df[10].values
y
```

```python
from sklearn.preprocessing import LabelBinarizer
label_encoder=LabelBinarizer()
y=label_encoder.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=50)
from sklearn.linear_model import LogisticRegression
```

```python
pen=['l1','l2']
c_val=2.**np.arange(-5,10,step=2)
c_val
```

```python
param_grid=[{'C':c_val,'penalty':pen}]
model=LogisticRegression(solver='liblinear')
from sklearn.model_selection import GridSearchCV
```

```python
cv_log=GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
cv_log
cv_log.fit(xtrain,ytrain[:,0])
```

```python
best_c=cv_log.best_params_['C']
best_penalty=cv_log.best_params_['penalty']
print(best_c)
print(best_penalty)
```

```python
classifier=LogisticRegression(C=best_c,penalty=best_penalty,solver='liblinear')
classifier.fit(xtrain,ytrain[:,0])
```

```python
prediction=classifier.predict(xtest)
y_proba=classifier.predict_proba(xtest)
print(ytest[:,0])
print(prediction)
print(y_proba)
print(classifier.score(xtest,ytest))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve,auc
fpr,tpr,threshold=roc_curve(ytest,y_proba[:,1])
plt.plot(fpr,tpr)
plt.show()
print(auc(fpr,tpr))
```
